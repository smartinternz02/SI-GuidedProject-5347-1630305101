```python
#Importing required libraries 
import re 
import numpy as np
import pandas as pd
import nltk
```

```python
#downloading the stopwords
nltk.download('stopwords')

#collection of the stopwords is corpus
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
data = []

from nltk.stem import WordNetLemmatizer

#create an object for wordnet lemmatizer
wordnet = WordNetLemmatizer()
```

```python
#reading the dataset
df = pd.read_csv("Reviews.csv")
```

```python
df
```

```python
#checking for null values
df.isnull().sum()
```

```python
#filling the null values 
df['body'] = df['body'].fillna('').apply (str)
df['name'] = df['name'].fillna('').apply (str)
df['title'] = df['title'].fillna('').apply (str)
df['helpfulVotes'] = df['helpfulVotes'].fillna('').apply (str)
```

```python
#rechecking the dataset after filling 
df.isnull().sum()
```

```python
#storing the rating values in a list
a= df['rating'].tolist()
a
```

```python
#assining 1 to rating no from 1 to 3(positive) and the rest as 0(negative) 
d =[]
for i in range(len(a)):
    if a[i] >=3:
        d.append(1)
    else:
        d.append(0)

d
```

```python
#joining columns, title and body
df['Review'] = df[['title','body']].agg(' '.join, axis =1)
df.head()
```

```python
#making the above list as a column called emotion

dt = pd.DataFrame(d, columns = ['emotion'])
dt.head()
```

```python
#concatenating the newly created column with the datatset
data1 = pd.concat([df,dt], axis = 1)
data1.head()
```

```python
data1.drop(columns = ['asin','name', 'rating', 'date','verified', 'title', 'body', 'helpfulVotes'], axis =1, inplace = True)
```

```python
data1
```

```python
#Check data1 shape
data1.shape
```

```python
x = data1.iloc[: ,0].values
y = data1.iloc[: ,1].values
```

```python
#initializing empty array to append clean text

corpus = []

#no of rowa to be cleaned
for i in range(len(x)):
    
    temp = data1['Review'][i]
    
    #replacing punctuations and numbers using re library
    temp = re.sub('[^a-zA-Z]', ' ', temp)
    
    temp = re.sub(r'\d+', ' ', temp)
    
    temp = temp.replace('[^\W\s\S]', ' ')
    
    #converting all text to lower case 
    temp = temp.lower()
    
    #spliting to array( default delimeter is " ")
    temp = temp.split()
    
    #looping for Lemmatizing each word in string array
    temp = [ps.stem(word) for word in temp if not word in set(
    stopwords.words('english'))]

    #rejoining the string
    temp = ' '.join(temp)

    #append each string to create array of clean text 
    corpus.append(temp)
```

```python
#tokenization or vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 2000)
#z contains vectorized data
z = cv.fit_transform(corpus).toarray()
```

```python
x = cv.fit_transform(corpus).toarray()
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
#Splitting the data to train and test set
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state = 0)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#importing keras libraries
import tensorflow
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense 

#building the model
model = Sequential() 
model.add(Dense(2000, kernel_initializer ='random_uniform', activation = "relu"))
model.add(Dense(4000, kernel_initializer ='random_uniform', activation = "relu"))
model.add(Dense(4000, kernel_initializer ='random_uniform', activation = "relu"))
model.add(Dense(1, kernel_initializer ='random_uniform', activation = "sigmoid"))

#to accelarate the gradient 
model.compile(optimizer = "rmsprop", loss = "binary_crossentropy", metrics = ["accuracy"])
model.fit(x_train, y_train, batch_size = 128, epochs = 10)
```

```python
model.summary() 
```
